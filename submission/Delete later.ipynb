```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns

matplotlib.rcParams["figure.figsize"]=(12,10)

sns.set()
```

```python
df = pd.read_csv("skewed data.csv")
df.head()
```

```python
def identify_skewness(df, ftr):    
    fig, ax = plt.subplots(figsize=(8,6))

    mean = df[ftr].mean()
    median = df[ftr].median()
    mode = df[ftr].mode().values[-1]

    sns.kdeplot(df[ftr])
    ax.axvline(mean, linestyle='--', color='red', label="MEAN")
    ax.axvline(median, linestyle='--', color='green', label="MEDIAN")
    ax.axvline(mode, linestyle='--', color='magenta', label="MODE")

    if mean<median and median<mode and mean<mode:
        title = "strongly left skewed"
    elif not mean<median and median<mode and mean<mode:
        title = "left skewed"
    elif mean>median and median>mode and mean>mode:
        title = "strongly right skewed"
    elif not mean>median and median>mode and mean>mode:
        title = "right skewed"

    ax.set_title(ftr + "~" + title)
    plt.legend()
    plt.show()
```

```python
identify_skewness(df, "feature_1")
```

```python
identify_skewness(df, "feature_2")
```
